```python
def geoEncoding(data,position):
    tmp = data[['start_lat','start_lon']]
    geohashList = []
    for i in tmp.values:
        geohashList.append(geohash.encode(i[0],i[1],position))
    data['geohash{}'.format(position)] = geohashList
    return data
```

```python
def geoEncoding_end(data,position):
    tmp = data[['end_lat','end_lon']]
    geohashList = []
    for i in tmp.values:
        geohashList.append(geohash.encode(i[0],i[1],position))
    data['geohash_end{}'.format(position)] = geohashList
    return data
```

```python
def transform_datetime(starttime):
    #2018èŠ‚å‡æ—¥
    """
    1æœˆä»½(1.1 å…ƒæ—¦)
    2æœˆä»½(2.15-2.21  æ˜¥èŠ‚    2.11ï¼Œ2.24 å‘¨æœ«åŠ ç­)
    3æœˆä»½(æ— )
    4æœˆä»½(4.5-4.7   æ¸…æ˜ŽèŠ‚   4.8 ï¼Œ4.28 å‘¨æœ«åŠ ç­)
         (4.29-4.30  åŠ³åŠ¨èŠ‚)
    5æœˆä»½(5.1         åŠ³åŠ¨èŠ‚)
    6æœˆä»½(6.16-6.18   ç«¯åˆèŠ‚) 
    7æœˆä»½(æ— )
    8æœˆä»½(æ— )"""
    holiday=['1.1','2.15','2.16','2.17','2.18','2.19','2.20','2.21','4.5','4.6','4.7','4.29','4.30','5.1','6.16','6.17','6.18']
    work_weekend=['2.11','2.24','4.8','4.28']
    w_list=[]
    m_list=[]
    h_list=[]
    ish_list=[]
    for start in starttime:
        formattime = datetime.strptime(start, "%Y-%m-%d %H:%M:%S")
        m = formattime.month  # mè¡¨ç¤ºæœˆä»½
        m_list.append(m)
        date=str(m)+'.'+str(formattime.day)
        w = datetime.weekday(formattime)  # wä»£è¡¨æ˜ŸæœŸå‡ ï¼ˆ0ä»£è¡¨æ˜ŸæœŸä¸€ï¼‰
        ish=''
        if w > 4 :#å‘¨æœ«
            if date in work_weekend:
                ish='workday'#ä¸Šç­æ—¥
            else:
                ish = 'freeday'#ä¼‘æ¯æ—¥
            w='weekend'
        else :#å·¥ä½œæ—¥
            if date in holiday:
                ish='freeday'
            else:
                ish = 'workday'
            w='weekday'
        w_list.append(w)
        ish_list.append(ish)
        h = formattime.hour  # hè¡¨ç¤ºå°æ—¶ï¼ˆ24å°æ—¶åˆ¶ï¼‰
        h_list.append(h)
    return w_list,m_list,h_list,ish_list
```

```python
from math import radians, atan, tan, sin, acos, cos
def getDistance(latA, lonA, latB, lonB):
    ra = 6378140  # radius of equator: meter
    rb = 6356755  # radius of polar: meter
    flatten = (ra - rb) / ra  # Partial rate of the earth
    # change angle to radians
    radLatA = radians(latA)
    radLonA = radians(lonA)
    radLatB = radians(latB)
    radLonB = radians(lonB)

    try:
        pA = atan(rb / ra * tan(radLatA))
        pB = atan(rb / ra * tan(radLatB))
        x = acos(sin(pA) * sin(pB) + cos(pA) * cos(pB) * cos(radLonA - radLonB))
        c1 = (sin(x) - x) * (sin(pA) + sin(pB)) ** 2 / cos(x / 2) ** 2
        c2 = (sin(x) + x) * (sin(pA) - sin(pB)) ** 2 / sin(x / 2) ** 2
        dr = flatten / 8 * (c1 - c2)
        distance = ra * (x + dr)
        return distance  # meter
    except:
        return 0.0000001
```
